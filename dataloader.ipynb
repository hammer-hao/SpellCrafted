```python
import json
from tqdm import tqdm
import pandas as pd
```

```python
path = 'all-cards.json'

with open(path, 'rb') as file:
    dict = json.load(file)
    
dict_eng=[]
for card in tqdm(dict):
    if card['lang']=='en':
        dict_eng.append(card)
len(dict_eng)
```

```python
df = pd.DataFrame(dict_eng)
df.columns
```

```python
params = ['name', 'mana_cost', 'cmc', 'type_line', 'oracle_text', 'power', 'toughness', 'colors', 'color_identity', 'keywords', 'rarity', 'flavor_text']
train_df=df[params]
train_df['oracle_text']=train_df['oracle_text'].fillna('')
train_df['flavor_text']=train_df['flavor_text'].fillna('')
train_df['keywords']="["+train_df['keywords'].apply(lambda x: ', '.join(x))+"]"
train_df.head(5)
```

```python
train_df['text']=train_df['name']+':\n'+train_df['type_line']+'\n'+train_df['oracle_text']+'\n'+train_df['flavor_text']
```

```python
train_df
```

```python
train_df.to_csv('mtg_data.csv')
import pickle
with open('mtgdata.pickle', 'wb') as file:
    pickle.dump(train_df, file)
```
